```python
#Transform input data
x_ = PolynomialFeatures(degree=2, include_bias=False).fit_transform(x.reshape(-1, 1))

#Create a model and fit it
poly = LinearRegression().fit(x_, y)

#Predict response
y_predicted = poly.predict(x_)

print(f"\npredicted response:\n{y_predicted}")

#graph
plt.scatter(x, y, c='blue')
plt.plot(x, y_predicted, c='green')
plt.xlabel('GDP (USD)')
plt.ylabel('Life satisfaction')
plt.title('Fitted model with degree 2')
```

### LINEAR REGRESSION WITH STATSMODEL

```python
import numpy as np
import statsmodels.api as sm

x = np.array([56755.72171242, 44178.04737774, 40991.80813814, 8814.00098681, 
              43585.51198178, 13574.17183072, 6175.8760297 ,  17829.69832237, 
              53254.85637009,  17522.23018625,  42784.69836164,  36638.18492916, 
              41086.72967373,  18167.77372717,  12706.89121489,  52564.42917946, 
              61995.42280258,  35776.79516181,  30230.22630213,  34524.46986093, 
              13774.60527391,  14258.22933451, 101376.49657434,   9616.64500569, 
              45175.23189338,  38615.99518491,  74355.51585756,  12578.49547344, 
              19242.3664711 ,  16310.988409  ,  20881.76692993,   5734.63362915, 
              25732.01836475,  51545.48360953,  82081.59716162,  11006.2497364 , 
              44974.83187718,  56839.38177423])
y = np.array([7.3, 7.1, 6.9, 6.4, 7.4, 6.5, 6.3, 6.7, 7.6, 5.7, 7.6, 6.5, 7.0, 
              5.4, 5.6, 7.5, 7.0, 7.2, 6.0, 5.9, 5.9, 5.9, 6.9, 6.5, 7.4, 7.3, 
              7.6, 6.1, 5.4, 6.2, 5.9, 4.7, 6.3, 7.3, 7.5, 5.5, 6.8, 6.9])

#for the y-intercept, call this fucntion:
x = sm.add_constant(x)

#first argument is output then the second one is input
model = sm.OLS(y, x)
results = model.fit()
print(results.summary())

#extract info from the table of summary
print(f"\ncoefficient of determination: {results.rsquared}")
print(f"\nadjusted coefficient of determination: {results.rsquared_adj}")
print(f"\nregression coefficients: {results.params}")

#predict responses
print(f"\npredicted response:\n{results.fittedvalues}")
print(f"\npredicted response:\n{results.predict(x)}")

#graph
fig = sm.graphics.plot_partregress_grid(results)
```
